```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

### Demographic Data

```python
hh_demographic = pd.read_csv("../data/raw/hh_demographic.csv")
print(f"Demographic data shape: {hh_demographic.shape}")
hh_demographic.head()
```

```python
hh_demographic["household_key"].nunique()
```

```python
# Figure size
plt.figure(figsize=(10,5))

# Define figure
ax = sns.histplot(x=hh_demographic["AGE_DESC"].sort_values())

# Get the patches (bars) from the histogram
for patch in ax.patches:
    # Get the count (height of the bar)
    count = int(patch.get_height())
    
    # Only place text if count > 0
    if count > 0:
        # Place text on top of each bar
        ax.text(
            patch.get_x() + patch.get_width() / 2,  # X position (center of bar)
            patch.get_height() + 0.3,               # Y position (just above bar)
            count,                                  # Text (the count)
            ha='center',                            # Horizontal alignment
            va='bottom',                            # Vertical alignment
            fontsize=10                             # Font size
        )

# Title
plt.title('Age Distribution')

# X-axis label
plt.xlabel("Age Group")

# Display figure
plt.show()
```

```python
# Map Marital code to more readable format
def marital_map(x):
    if x=="A":
        return "Married"
    elif x=="B":
        return "Single"
    return "Unknown"

# Figure size
plt.figure(figsize=(10,5))

# Define figure
ax = sns.histplot(x=hh_demographic["MARITAL_STATUS_CODE"].apply(marital_map).sort_values())

# Get the patches (bars) from the histogram
for patch in ax.patches:
    # Get the count (height of the bar)
    count = int(patch.get_height())
    
    # Only place text if count > 0
    if count > 0:
        # Place text on top of each bar
        ax.text(
            patch.get_x() + patch.get_width() / 2,  # X position (center of bar)
            patch.get_height() + 0.3,               # Y position (just above bar)
            count,                                  # Text (the count)
            ha='center',                            # Horizontal alignment
            va='bottom',                            # Vertical alignment
            fontsize=10                             # Font size
        )

# Title
plt.title('Marital Distribution')

# X-axis label
plt.xlabel("Marital Status")

# Display figure
plt.show()
```

```python
# Dictionary for sorting income categories
income_sort = {
    'Under 15K': 0,
    '15-24K': 1,
    '25-34K': 2,
    '35-49K': 3,
    '50-74K': 4,
    '75-99K': 5,
    '100-124K': 6,
    '125-149K': 7,
    '150-174K': 8,
    '175-199K': 9,
    '200-249K': 10,
    '250K+': 11
}

# Figure size
plt.figure(figsize=(13,5))

# Define figure
ax = sns.histplot(x=hh_demographic["INCOME_DESC"].sort_values(key=lambda x : x.map(income_sort)))

# Get the patches (bars) from the histogram
for patch in ax.patches:
    # Get the count (height of the bar)
    count = int(patch.get_height())
    
    # Only place text if count > 0
    if count > 0:
        # Place text on top of each bar
        ax.text(
            patch.get_x() + patch.get_width() / 2,  # X position (center of bar)
            patch.get_height() + 0.3,               # Y position (just above bar)
            count,                                  # Text (the count)
            ha='center',                            # Horizontal alignment
            va='bottom',                            # Vertical alignment
            fontsize=10                             # Font size
        )

# Title
plt.title('Income Distribution')

# X-axis label
plt.xlabel("Income Group")

# Display figure
plt.show()
```

### Campaign Data

```python
campaign_table = pd.read_csv("../data/raw/campaign_table.csv")
print(f"Campaign Table shape: {campaign_table.shape}")
campaign_table.head()
```

```python
# Figure size
plt.figure(figsize=(10,7))

# Define figure
ax = sns.barplot(
    data = campaign_table.groupby("DESCRIPTION",as_index=False)["CAMPAIGN"].nunique(),
    x="DESCRIPTION",
    y="CAMPAIGN"
    )

# Get the patches (bars) from the histogram
for patch in ax.patches:
    # Get the count (height of the bar)
    count = int(patch.get_height())
    
    # Only place text if count > 0
    if count > 0:
        # Place text on top of each bar
        ax.text(
            patch.get_x() + patch.get_width() / 2,  # X position (center of bar)
            patch.get_height() + 0.3,               # Y position (just above bar)
            count,                                  # Text (the count)
            ha='center',                            # Horizontal alignment
            va='bottom',                            # Vertical alignment
            fontsize=10                             # Font size
        )

# Title
plt.title('Campaign Type Distribution')

# X-axis and Y-axis label
plt.xlabel("Campaign Type")
plt.ylabel("Campaign count")

# Display figure
plt.show()
```

```python
campaign_desc = pd.read_csv("../data/raw/campaign_desc.csv")
print(f"Campaign Desc shape: {campaign_desc.shape}")
campaign_desc.head()
```

```python
campaign_desc["CAMPAIGN_DURATION"] = campaign_desc["END_DAY"] - campaign_desc["START_DAY"]
```

```python
# Figure size
plt.figure(figsize=(13,7))

# Define figure
ax = sns.barplot(
    x=campaign_desc["CAMPAIGN"],
    y=campaign_desc["CAMPAIGN_DURATION"]
    )

# Get the patches (bars) from the histogram
for patch in ax.patches:
    # Get the count (height of the bar)
    count = int(patch.get_height())
    
    # Only place text if count > 0
    if count > 0:
        # Place text on top of each bar
        ax.text(
            patch.get_x() + patch.get_width() / 2,  # X position (center of bar)
            patch.get_height() + 0.3,               # Y position (just above bar)
            count,                                  # Text (the count)
            ha='center',                            # Horizontal alignment
            va='bottom',                            # Vertical alignment
            fontsize=10                             # Font size
        )

# Title
plt.title('Campaign Duration Distribution')

# X-axis and Y-axis label
plt.xlabel("Campaign ID")
plt.ylabel("Campaign Duration")

# Median campaign duration
ax.axhline(y=campaign_desc["CAMPAIGN_DURATION"].median(),
           color='r',
           linestyle='--',
           linewidth=1.5,
           label=f"Median Campaign Duration: {campaign_desc['CAMPAIGN_DURATION'].median():.2f}")
plt.legend()

# Display figure
plt.show()
```
