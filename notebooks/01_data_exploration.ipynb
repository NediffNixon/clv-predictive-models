```python
# Figure size
plt.figure(figsize=(10,5))

# Define figure
ax = sns.histplot(x=hh_demographic["AGE_DESC"].sort_values())

# Get the patches (bars) from the histogram
for patch in ax.patches:
    # Get the count (height of the bar)
    count = int(patch.get_height())
    
    # Only place text if count > 0
    if count > 0:
        # Place text on top of each bar
        ax.text(
            patch.get_x() + patch.get_width() / 2,  # X position (center of bar)
            patch.get_height() + 0.3,               # Y position (just above bar)
            count,                                  # Text (the count)
            ha='center',                            # Horizontal alignment
            va='bottom',                            # Vertical alignment
            fontsize=10                             # Font size
        )

# Title
plt.title('Age Distribution')

# X-axis label
plt.xlabel("Age Group")

# Display figure
plt.show()
```

```python
coupon_sent=coupon.groupby("CAMPAIGN",as_index=False)["COUPON_UPC"].nunique()
coupon_redeemed=coupon_redempt.groupby("CAMPAIGN",as_index=False)["COUPON_UPC"].nunique()

print(f"Coupons sent vs Coupons redeemed linear correlation: {coupon_sent['COUPON_UPC'].corr(coupon_redeemed['COUPON_UPC']):.2f}")
```

Can confirm that coupons sent has high positive linear correlation with coupons redeemed. But what proportion of coupons sent is getting converted?

```python
# Figure size
plt.figure(figsize=(15,7))

plot_df = pd.DataFrame({
    'CAMPAIGN': pd.concat([coupon_sent['CAMPAIGN'],coupon_redeemed['CAMPAIGN']]),
    'COUPON_UPC': pd.concat([coupon_sent['COUPON_UPC'],coupon_redeemed['COUPON_UPC']]),
    'Group': ['Coupons sent']*30 + ['Coupons redeemed']*30
})
# Define figure
ax = sns.barplot(
    x=plot_df['CAMPAIGN'],
    y=plot_df['COUPON_UPC'],
    hue=plot_df['Group']
    )

# Get the patches (bars) from the histogram
for patch in ax.patches:
    # Get the count (height of the bar)
    count = int(patch.get_height())
    
    # Only place text if count > 0
    if count > 0:
        # Place text on top of each bar
        ax.text(
            patch.get_x() + patch.get_width() / 2,  # X position (center of bar)
            patch.get_height() + 0.3,               # Y position (just above bar)
            count,                                  # Text (the count)
            ha='center',                            # Horizontal alignment
            va='bottom',                            # Vertical alignment
            fontsize=10                             # Font size
        )

# Title
plt.title('Campaign Conversion Distribution') 

# X-axis and Y-axis label
plt.xlabel("Campaign ID")
plt.ylabel("Coupons")

# Display figure
plt.show()
```

```python
# Rename columns for join
coupon_sent = coupon_sent.rename({"COUPON_UPC":"COUPON_SENT"},axis=1)
coupon_redeemed = coupon_redeemed.rename({"COUPON_UPC":"COUPON_REDEEMED"},axis=1)

# Join Coupons sent and Coupons redeemed
coupon_stat = coupon_sent.merge(coupon_redeemed,on="CAMPAIGN",how="left")

# Calculate conversion rate
coupon_stat["COUPON_CONVERSION_RATE"] = np.round(coupon_stat["COUPON_REDEEMED"]/coupon_stat["COUPON_SENT"],2)

# Print a sorted (by conversion rate) dataframe
coupon_stat.sort_values("COUPON_CONVERSION_RATE",ascending=False).reset_index(drop=True)
```

```python
# Overall conversion rate. Weighted conversion rate to account for campaign size.
print(f"Overall conversion rate: {np.sum(coupon_stat['COUPON_REDEEMED'])/np.sum(coupon_stat['COUPON_SENT']):.2f}")
```

```python
transactions = pd.read_csv("../data/raw/transaction_data.csv")
print(f"transaction data shape: {transactions.shape}")
transactions.head()
```

```python
# Days recorded in transaction data - 710 days
transactions["DAY"].agg(["min","max"])
```

```python
# Figure size
plt.figure(figsize=(15,7))

# Define figure
data = transactions.groupby("DAY",as_index=False)["BASKET_ID"].nunique()
sns.barplot(x=data["DAY"],
            y=data["BASKET_ID"]
            )

# Title
plt.title('Daily Purchases') 

# X-axis and Y-axis label
plt.xlabel("DAY")
plt.ylabel("VISITS")

# Get current axis
ax = plt.gca()

# Assuming your DAY column contains numerical values
# Get the min and max of your day values
days = transactions['DAY'].unique()
min_day = min(days)
max_day = max(days)

# Create labels at intervals of 100
tick_labels = list(range(0, max_day + 100, 100))
if tick_labels[-1] > max_day:
    tick_labels = tick_labels[:-1]  # Remove last label if it exceeds max day

# Calculate positions for these labels (map them to bar positions)
positions = []
for label in tick_labels:
    # Find the closest day to each label value
    closest_day_idx = min(range(len(days)), key=lambda i: abs(days[i] - label))
    positions.append(closest_day_idx)

# Set the ticks, labels and horizonal median line
ax.set_xticks(positions)
ax.set_xticklabels(tick_labels)
ax.axhline(np.median(data["BASKET_ID"]),color='r',
           linestyle='--',
           linewidth=1.5,
           label=f"Median cut-off: {np.median(data['BASKET_ID']):.2f}")

# Display figure
plt.legend()
plt.show()
```

Seems like days before 101 are under-representative of the whole trend. Scenerio of data drift. So we will use transaction data from day no. 101 because they are the best representation of the overall purchase behaviour.

```python
transactions_selected = transactions.loc[transactions["DAY"]>=101].reset_index(drop=True)
```
